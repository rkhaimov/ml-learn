```python
import sklearn.neural_network as nn
import sklearn.datasets as ds
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

clf = nn.MLPClassifier(hidden_layer_sizes=(32, 32), learning_rate="adaptive",
                       solver="lbfgs", max_iter=2000, activation="relu")
```

```python
X, y = ds.make_moons(n_samples=100, noise=0.2)

plt.scatter(X[:, 0], X[:, 1], c=y)
plt.colorbar()
```

```python
clf.fit(X, y)
```

```python
linspace = np.linspace(-2, 2, 10)
left, right = np.meshgrid(linspace, linspace)
xs = np.column_stack([left.ravel(), right.ravel()])

ones = clf.predict_proba(xs)[:, 1]
z = ones.reshape([10, 10])

plt.contourf(left, right, z, cmap="BuGn")
plt.colorbar()
plt.scatter(X[:, 0], X[:, 1], c=y)

clf.score(X, y)
```
