```python
from sklearn.neural_network import MLPClassifier
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

X = np.array([[0, 0], [1, 1], [1, 0], [0, 1]])
y = np.array([1, 1, 0, 0])

clf = MLPClassifier(hidden_layer_sizes=(16, 4, 6), learning_rate="adaptive", max_iter=2_000)

clf.fit(X, y)
```

```python
linspace = np.linspace(0, 1, 50)
left, right = np.meshgrid(linspace, linspace)
x_exam = np.column_stack([left.reshape([-1, 1]), right.reshape([-1, 1])])

one_proba = clf.predict_proba(x_exam)[:, 1]

plt.scatter(x_exam[:, 0], x_exam[:, 1], c=one_proba, cmap="BuGn")
plt.colorbar()

clf.score(X, y)
```
