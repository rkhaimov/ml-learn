```python
import sklearn.neural_network as nn
import sklearn.model_selection as ms
import sklearn.datasets as ds
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

digits = ds.load_digits()

X_train, X, y_train, y = ms.train_test_split(digits.data, digits.target, test_size=0.4)
X_test, X_cv, y_test, y_cv = ms.train_test_split(X, y, test_size=0.5)

plt.imshow(X_train[0].reshape([8, 8]), cmap="Greys_r")
```

```python
history = []

for l in [0.001, 0.01, 0.1, 0.3, 0.6, 0.8, 1, 3, 6, 10, 15, 20]:
    clf = nn.MLPClassifier(hidden_layer_sizes=2 ** 8, alpha=l)

    clf.fit(X_train, y_train)

    history.append([l, clf.score(X_test, y_test), clf.score(X_train, y_train), clf.score(X_cv, y_cv)])
```

```python
nhistory = np.array(history)

plt.plot(nhistory[:, 0], 1 - nhistory[:, 1], label='Test')
plt.plot(nhistory[:, 0], 1 - nhistory[:, 2], label='Train')
plt.plot(nhistory[:, 0], 1 - nhistory[:, 3], label='CV')

plt.xlabel('alpha')
plt.ylabel('error')
plt.legend()
```
